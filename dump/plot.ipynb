```python
import json
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
with open('base_popularity.json', 'r') as json_file:
    base_popularity = json.load(json_file)
with open('base_temprature.json', 'r') as json_file:
    base_temprature = json.load(json_file)

```

```python
with open('lora_temprature.json', 'r') as json_file:
    lora_temprature = json.load(json_file)

with open('lora_popularity.json', 'r') as json_file:
    lora_popularity = json.load(json_file)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)

sns.regplot(x=np.arange(100), y= list(reversed(list(base_temprature.values()))), label='base')
sns.regplot(x=np.arange(100), y= list(reversed(list(lora_temprature.values()))), label='lora')
plt.xlabel('temprature')
plt.ylabel('NDCG@10')
plt.legend()

plt.subplot(1,2,2)

sns.regplot(x=np.arange(100), y= list(base_popularity.values()), label='base')
sns.regplot(x=np.arange(100), y= list(lora_popularity.values()), label='lora')
plt.xlabel('popularity')
plt.ylabel('NDCG@10')
plt.legend()
```
